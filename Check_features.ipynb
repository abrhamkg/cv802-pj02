```python
import pycolmap
import cv2
import matplotlib.pyplot as plt

# Load the database saved by api.py after feature extraction and matching
database_path="/Users/daromog/Desktop/David/PhD/Classes/Advanced_3D_CV/CV802-starter-code_drm_SFM/assignment1/test_images/colmap/database.db"
db=pycolmap.Database(database_path) 
#Read specific camera points 1-11
keypoints=db.read_keypoints(1) #Format: [x, y, scale, orientation]

#Read image and plot keypoints
image = cv2.imread(f'/Users/daromog/Desktop/David/PhD/Classes/Advanced_3D_CV/CV802-starter-code_drm_SFM/assignment1/test_images/images/DSC_0504.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 10))
plt.imshow(image_rgb)
plt.scatter(keypoints[:, 0], keypoints[:, 1], s=1, c='r', marker='o')   
plt.title('Keypoints-SIFT')
plt.show()
```
